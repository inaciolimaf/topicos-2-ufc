```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
resultadoTempo = pd.read_excel("resultadoTempo.xlsx", index_col="Unnamed: 0").T
resultadoMem = pd.read_excel("resultadoMemoria.xlsx", index_col="Unnamed: 0").T
```

```python
resultadoTempo = resultadoTempo.reset_index().rename(columns={'index': 'Tamanho'})
resultadoMem = resultadoMem.reset_index().rename(columns={'index': 'Tamanho'})
```

```python
display(resultadoTempo)
sns.set_style("darkgrid") 
plt.figure(figsize=(100, 60))
plt.tick_params(labelsize=70)
for i in range(0, 2):
    plot = sns.lineplot(x="Tamanho", y=i, data=resultadoTempo, linewidth=10)
    
labels = {"Maior Valor V1": 20, "Maior Valor V2": 20}
for i, (key, value) in enumerate(labels.items()):
    plt.annotate(key, (resultadoTempo["Tamanho"][value], resultadoTempo[i][value]), fontsize=50,
                 color='black', fontweight='bold', rotation=15)
plt.xscale("log")
plot.set_title("Tempo nos algoritmos", fontsize=120)  # Ttulo
plot.set_xlabel("Tamanho do vetor", fontsize=80)  # Rtulo do eixo x
plot.set_ylabel("Tempo(s)", fontsize=80)  # Rtulo do eixo y

plt.show()
```
